## KNN (lähima naabri meetod)

Esimese klassifitseerimisalgoritmina vaatame lähima naabri meetodit (KNN). 

Selle kohta on hästi palju materjali, kuid üldine algoritmikirjeldus on järgmine:

1) Sul on mingid andmepunktid (nt sa tead iga inimese kohta tema vanust, pikkust ja sugu)

2) Sul on osade andmete kohta rohkem informatsiooni (nt sa tead osade inimeste juustevärvi, aga teise rühma puhul info puudub).

3) Sul on vaja teada saada ka teiste inimeste juuksevärv.

4) Sa arvutad inimestevahelised kaugused

5) Võtad n-lähimat ning teed selle põhjal otsuse.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid') # Optional styling
plt.rcParams.update({'font.size': 14}) # Change font size

data = pd.DataFrame(
    {'x':[1,3,2,3,1,4,9,5,6,6,9],
    'y':[7,9,4,3,8,1,3,6,4,2,9],
    'label':  ['Kass']*5 + ['Koer']*6})


unlabelled_test_instance = [3, 4.5]

def plot_points(data):
    
    # Specify figure parameters
    fig, ax = plt.subplots(figsize=(8, 8)) 
    ax.set_aspect('equal')
    
    # Data to plot
    ax.plot(data.loc[data.label == 'Kass', 'x'], data.loc[data.label == 'Kass', 'y'], 'bo', label='Kassid') 
    ax.plot(data.loc[data.label == 'Koer', 'x'], data.loc[data.label == 'Koer', 'y'], 'ro', label='Koerad')

    # Axis labels
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    
    # Specify axes parameters
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    
    plt.legend(markerscale=1, frameon=True, loc='lower right', facecolor='white')

    for i, label in enumerate(data.label):
        # TODO: add labels to each data point
        x =data.iloc[i]["x"] +0.1
        y = data.iloc[i]["y"] +0.1
        plt.annotate(label,[x,y]) 
    
    return ax

plot_points(data) # uncomment to visualize the dataset

# TODO
plt.plot(3,4.5,'co')
plt.annotate("???",[3.1,4.5])
plt.show()
```
